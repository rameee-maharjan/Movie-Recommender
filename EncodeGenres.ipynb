```python
import sklearn
import numpy as np
import pandas as pd
import random
import math
import pickle

```

```python
# find distinct gentres in different movies
def find_genres(movie_ratings):
    # listing all genres from each row
    genres = movie_ratings['genres'].str.split("|")
    all_genres = []
    for genre in genres:
        all_genres.extend(genre)

    # taking only unique genres from all_genres listed
    distinct_genres = set(all_genres)
    return distinct_genres
```

```python
movie_ratings = pd.read_csv("./datasets/movie_ratings.csv")
```

```python
genre_list = list(find_genres(movie_ratings))
genre_list
```

```python
print(type(genre_list))
```

```python
# Function to find ONE HOT ENCODING
#genre_list = ['Action', 'Adventure', 'Animation','Children','Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

def one_hot_encoding(genres):
    genres = genres.strip().split("|")
    OHE=[0]*len(genre_list)
    for i in range(len(genre_list)):
        # place 1 if the movie has that genre
        if genre_list[i] in genres:
            OHE[i] = 1
            
    return OHE

```

```python

OHE_df = pd.DataFrame(columns = genre_list )
# find OHE for every movie in movie ratings
for i, row in movie_ratings.iterrows():
    OHE = [one_hot_encoding(row["genres"])]
    OHE_df = OHE_df.append(pd.DataFrame( OHE, 
               columns=genre_list),
               ignore_index = True)
```

```python
# combine the encoding for genres to movie ratings
mv_df = movie_ratings.copy()
data = mv_df.join(OHE_df)
```

```python
# Rearrange columns to place rating to the end of the dataframe
col = ['userId', 'movieId', 'title', 'genres', 'year', 'Action',
       'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'rating']
data = data[col]
data
```
