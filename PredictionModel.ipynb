```python
import pandas as pd
import math
from sklearn import linear_model
import os
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from joblib import dump, load
```

```python
data = pd.read_csv(".\datasets\Data_OHE.csv", index_col = 0)
```

```python
data.head()
```

# Features selection for using as input attributes and output in regression

```python
X = data[['year', 'Action', 'Adventure', 'Animation','Children','Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
y = data [['rating']]
```

```python
# Splitting into training and testing dataset for regression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_test
```

# Linear Regression

```python
# FIt the model
reg_model = LinearRegression().fit(X_train, y_train)
```

```python
#Save the model
if not os.path.isfile("./model/reg_pred_model.joblib"):
    dump(reg_model, './model/reg_pred_model.joblib')
```

```python
# Find model accuracy
accuracy = round(reg_model.score(X_test, y_test)* 100, 2)
print(accuracy)
```

```python
y_pred = reg_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("MSE = ", MSE, "RMSE= ", RMSE)
```

# Ridge regression

```python
# FIt the model
ridge_model = linear_model.Ridge().fit(X_train, y_train)
#Save the model
if not os.path.isfile("./model/ridge_pred_model.joblib"):
    dump(ridge_model, './model/ridge_pred_model.joblib')
# Find model accuracy
accuracy = round(ridge_model.score(X_test, y_test)* 100, 2)
print(accuracy)

y_pred = ridge_model.predict(X_test)
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("MSE = ", MSE, "RMSE= ", RMSE)

```

# Lasso regression

```python
# FIt the model
lasso_model = linear_model.Lasso(alpha=0.1).fit(X_train, y_train)
#Save the model
if not os.path.isfile("./model/lasso_pred_model.joblib"):
    dump(lasso_model, './model/lasso_pred_model.joblib')
# Find model accuracy
accuracy = round(lasso_model.score(X_test, y_test)* 100, 2)
print(accuracy)


y_pred = lasso_model.predict(X_test)
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("MSE = ", MSE, "RMSE= ", RMSE)
```
